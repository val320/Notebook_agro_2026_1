### Ejemplos de Código (Sencillos y Útiles)

#### 1. Cálculo del Índice de Madurez

En la industria de frutas, el índice de madurez ($IM$) es clave para decidir el momento de cosecha o procesamiento. Se calcula como la relación entre Sólidos Solubles Totales ($SST$) y la Acidez Titulable ($AT$).

```
# Datos de laboratorio
sst = 14.5  # Grados Brix
acidez = 0.8 # Porcentaje de ácido cítrico

indice_madurez = sst / acidez

print(f"El Índice de Madurez es: {indice_madurez:.2f}")

if indice_madurez > 15:
    print("Estado: Fruta lista para procesamiento de dulcería.")
else:
    print("Estado: Fruta requiere más tiempo de maduración.")
```

#### 2. Monitoreo de Humedad en Almacén

Imagina que recibes datos de un sensor. Este código evalúa si las condiciones de almacenamiento son seguras para evitar hongos.

```
humedad_actual = 72  # Porcentaje de humedad detectado

limite_seguro = 65

if humedad_actual > limite_seguro:
    diferencia = humedad_actual - limite_seguro
    print(f"ALERTA: Humedad excedida por {diferencia}%. Activar extractores.")
else:
    print("Condiciones de almacenamiento óptimas.")
```

#### 3. Conversión de Unidades de Producción

A menudo se necesita calcular el rendimiento de una materia prima tras el proceso de extracción (por ejemplo, aceite o jugo).

```
def calcular_rendimiento(peso_inicial, peso_final):
    rendimiento = (peso_final / peso_inicial) * 100
    return rendimiento

# Ejemplo: Extracción de aceite de palma
fruto_fresco = 500  # kg
aceite_obtenido = 110 # kg

resultado = calcular_rendimiento(fruto_fresco, aceite_obtenido)

print(f"El rendimiento de la extracción es del {resultado}%")
```
