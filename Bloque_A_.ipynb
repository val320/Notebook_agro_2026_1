**Ejercicio 1:**
# Capacidad de Silos de Granos
**Problema:**

Se requiere calcular cuántas toneladas de maíz puede almacenar un silo cilíndrico.

**Requerimientos:**

El usuario ingresa el radio y la altura del silo en metros.

**Datos:**


**Densidad del maíz ≈0.75 ton/m3**
**Volumen V=π*r2*h**

**Prueba:**

Radio = 3m, Altura = 10m.

**Salida** ≈212.06 toneladas.

```python
# ENTRADA: Ingresar El radio
# Ingresar la altura en metros(m)
radio = float(input("Ingrese el radio del silo en metros: "))
altura = float(input("Ingrese la altura del silo en metros: "))

# PROCESO: Calcular el volumen
# Densidad del maíz = 0.75 ton/m3
# Volumen V=π*r2*h
# SALIDA: cuántas toneladas de maíz puede almacenar un silo cilíndrico.
volumen = 3.14159 * (radio ** 2) * altura
toneladas = volumen * 0.75
print("El silo puede almacenar", toneladas, "toneladas de maíz.")
```

El código realiza el siguiente proceso para calcular la capacidad de un silo:

*   `# ENTRADA: Ingresar El radio` y `# Ingresar la altura en metros(m)`: Estas son líneas de comentarios que indican qué se espera como entrada del usuario.

*   `radio = float(input("Ingrese el radio del silo en metros: "))`: Esta línea pide al usuario que ingrese el radio del silo. El valor ingresado se convierte a un número de punto flotante (`float`) y se almacena en la variable `radio`.

*   `altura = float(input("Ingrese la altura del silo en metros: "))`: Similar a la anterior, esta línea pide al usuario la altura del silo, la convierte a `float` y la guarda en la variable `altura`.

*   `# PROCESO: Calcular el volumen`, `# Densidad del maíz = 0.75 ton/m3`, y `# Volumen V=π*r2*h`: Estos son comentarios que describen los cálculos a realizar y las fórmulas utilizadas.

*   `# SALIDA: cuántas toneladas de maíz puede almacenar un silo cilíndrico.`: Otro comentario que indica cuál será el resultado final del programa.

*   `volumen = 3.14159 * (radio ** 2) * altura`: Esta es la línea clave para el cálculo. Utiliza la fórmula del volumen de un cilindro (`π * r² * h`), donde `3.14159` es una aproximación de Pi, `radio ** 2` calcula el radio al cuadrado, y el resultado se multiplica por la altura. El valor se guarda en la variable `volumen`.

*   `toneladas = volumen * 0.75`: Aquí se calcula la cantidad de toneladas de maíz que puede almacenar el silo. El volumen calculado se multiplica por la densidad del maíz (`0.75 ton/m3`). El resultado se guarda en la variable `toneladas`.

*   `print("El silo puede almacenar", toneladas, "toneladas de maíz.")`: Finalmente, esta línea imprime un mensaje en la consola mostrando el resultado, es decir, cuántas toneladas de maíz puede almacenar el silo.

**Ejercicio 2:**

**Dosificación de Fertilizante Líquido**

**Problema:**

Un sistema de riego por goteo requiere aplicar una dosis de fertilizante por hectárea.

**Requerimientos:**

El usuario ingresa las hectáreas totales y la dosis requerida (litros/hectárea).

**Datos:**

El fertilizante se diluye en agua en una relación 1:500 (1 litro de producto por 500 de agua).

**Prueba:**

5 hectáreas, Dosis 2 L/ha.

**Salida:**

10L de producto y 5000L de agua total.

```python
# ENTRDA: Ingresar las hectareas totales (h)
# La dosis requerida (lt/h)
h = float(input("Ingrese las hectareas totales: "))
lt = float(input("Ingrese la dosis requerida (lt/ha): "))

# PROCESO:
# Cantidad total de producto (litros)
dosis_total = h * lt

# Relación 1:500 → por cada 1 litro de producto, 500 litros de agua
agua_total = dosis_total * 500

# SALIDA:
print("Dosis total usada (lt):", dosis_total)
print("Cantidad de agua total (lt):", agua_total)
```

El código del segundo ejercicio realiza el siguiente proceso para calcular la dosificación de fertilizante:

*   `# ENTRDA: Ingresar las hectareas totales (h)` y `# La dosis requerida (lt/h)`: Estas son líneas de comentarios que indican qué se espera como entrada del usuario.

*   `h = float(input("Ingrese las hectareas totales: "))`: Esta línea pide al usuario que ingrese el número total de hectáreas. El valor ingresado se convierte a un número de punto flotante (`float`) y se almacena en la variable `h`.

*   `lt = float(input("Ingrese la dosis requerida (lt/ha): "))`: Similar a la anterior, esta línea pide al usuario la dosis requerida en litros por hectárea, la convierte a `float` y la guarda en la variable `lt`.

*   `# PROCESO:`: Un comentario que indica el inicio de la sección de procesamiento.

*   `# Cantidad total de producto (litros)`: Un comentario que describe el cálculo siguiente.

*   `dosis_total = h * lt`: Esta línea calcula la cantidad total de fertilizante líquido (producto) necesaria, multiplicando las hectáreas totales (`h`) por la dosis requerida por hectárea (`lt`). El resultado se guarda en la variable `dosis_total`.

*   `# Relación 1:500 → por cada 1 litro de producto, 500 litros de agua`: Un comentario que explica la relación de dilución del fertilizante.

*   `agua_total = dosis_total * 500`: Aquí se calcula la cantidad total de agua necesaria. Se multiplica la `dosis_total` de fertilizante por 500, ya que la relación es 1 litro de producto por 500 litros de agua. El resultado se guarda en la variable `agua_total`.

*   `# SALIDA:`: Un comentario que indica el inicio de la sección de salida.

*   `print("Dosis total usada (lt):", dosis_total)`: Esta línea imprime un mensaje en la consola mostrando la cantidad total de fertilizante (producto) que se utilizará, contenida en la variable `dosis_total`.

*   `print("Cantidad de agua total (lt):", agua_total)`: Finalmente, esta línea imprime un mensaje en la consola mostrando la cantidad total de agua necesaria para la dilución, contenida en la variable `agua_total`.

**Ejercicio 3:**

**Rendimiento de Pulpa en Frutas**


**Problema:**

Una planta procesadora necesita saber el porcentaje de aprovechamiento.

**Requerimientos:**


Se ingresa el peso total de la fruta comprada y el peso de los residuos (cáscaras y semillas).

**Datos:**

Pulpa=PT−R

R=(Pulpa/PT)*100


**Prueba:**

PT = 500kg, Residuos = 150kg.

**Salida:** 350kg de pulpa (70%).

```python
# ENTRADA DE DATOS
pt = float(input("Ingrese el peso total de la fruta (kg): "))
residuos = float(input("Ingrese el peso de los residuos (kg): "))

# PROCESO
# Cálculo de la pulpa obtenida
pulpa = pt - residuos

# Cálculo del rendimiento en porcentaje
rendimiento = (pulpa / pt) * 100

# SALIDA DE RESULTADOS
print("\n----- RESULTADOS -----")
print("Pulpa obtenida:", pulpa, "kg")
print("Rendimiento:", rendimiento, "%")
```

El código del tercer ejercicio realiza el siguiente proceso para calcular el rendimiento de pulpa en frutas:

*   `# ENTRADA DE DATOS`: Es un comentario que indica la sección donde se solicitan los datos al usuario.

*   `pt = float(input("Ingrese el peso total de la fruta (kg): "))`: Esta línea pide al usuario que ingrese el peso total de la fruta en kilogramos. El valor se convierte a un número de punto flotante (`float`) y se guarda en la variable `pt` (peso total).

*   `residuos = float(input("Ingrese el peso de los residuos (kg): "))`: Similar a la anterior, esta línea pide el peso de los residuos (cáscaras y semillas), lo convierte a `float` y lo almacena en la variable `residuos`.

*   `# PROCESO`: Un comentario que marca el inicio de la sección de cálculos.

*   `# Cálculo de la pulpa obtenida`: Un comentario que describe el cálculo siguiente.

*   `pulpa = pt - residuos`: Esta línea calcula la cantidad de pulpa obtenida restando el peso de los `residuos` al `pt` (peso total de la fruta). El resultado se almacena en la variable `pulpa`.

*   `# Cálculo del rendimiento en porcentaje`: Un comentario que describe el cálculo siguiente.

*   `rendimiento = (pulpa / pt) * 100`: Aquí se calcula el rendimiento porcentual de la pulpa. Se divide la `pulpa` obtenida por el `pt` (peso total) y se multiplica por 100 para expresarlo como porcentaje. El resultado se guarda en la variable `rendimiento`.

*   `# SALIDA DE RESULTADOS`: Un comentario que indica la sección de los resultados finales.

*   `print("\n----- RESULTADOS -----")`: Imprime una línea separadora para mejorar la presentación de los resultados en la consola.

*   `print("Pulpa obtenida:", pulpa, "kg")`: Esta línea imprime la cantidad de pulpa obtenida, mostrando el valor de la variable `pulpa` en kilogramos.

*   `print("Rendimiento:", rendimiento, "%")`: Finalmente, esta línea imprime el rendimiento porcentual, mostrando el valor de la variable `rendimiento`.
